## Plotting a Line

Plot a line starting at **x=4, y=0** and ending at **x=20, y=4**. The function you should use is [pyplot.plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)

```python
# YOUR CODE HERE
plt.figure(figsize=(12,4))
x_data = [4,20]
y_data = [0,4]
plt.plot(x_data, y_data, linestyle="--")
```

## Plotting Multiple Curves

Plot 3 different curves on the same graph, still with pyplot.plot(). 

```python
# YOUR CODE HERE
import numpy as np
data_my = np.arange(0, 10, 0.2)
x = np.sin(data_my)
y = np.cos(data_my)
z = np.tan(data_my)
plt.plot(data_my, x, color='r', label='sin') 
plt.plot(data_my, y, color='b', label='cos') 
plt.plot(data_my, z, color='g', label='tan')
plt.legend()
plt.xlabel("Angles")
plt.ylabel("Values")
plt.show()
```
